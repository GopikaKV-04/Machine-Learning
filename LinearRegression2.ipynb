```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv('salary.csv')
df.head()
```

```python
plt.scatter(df['years_experience'],df['salary'])
plt.xlabel("years_experience")
plt.ylabel("salary")
```

```python
##seaborn for visualisation
import seaborn as sns
sns.pairplot(df)
```

```python
x=df['years_experience']
x
##this independant variabe is inside single[] which is a 1-D array 
```

```python
type(x)
##type is series when you put x in single[]
##make sure you put [[]] for ur independant variable as it creates a dataframe (two dimensional array)
```

```python
x=df[['years_experience']]
x
```

```python
type(x)
```

```python
np.array(x).shape
```

```python
y=df['salary']
y
##dependant variable should be in series(1-D format)
```

```python
##train tet split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(x_train,y_train)
```

```python
##best fit line
plt.plot(x_train,model.predict(x_train))
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,model.predict(x_train))
```

```python
y_pred=model.predict(x_test)
```
